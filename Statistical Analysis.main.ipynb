```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('F:\StatsProj\Spotifynewclean.csv')
df.head()
```

```python
plt.figure(figsize=(15,6))
plt.title('Correlation Matrix', fontsize=25)
sns.heatmap(df.corr(), cmap='RdBu', annot=True)
plt.show()
```

Definition: correlation matrix is simply a table which displays the correlation coefficients for different variables. The matrix depicts the correlation between all the possible pairs of values in a table. It is a powerful tool to summarize a large dataset and to identify and visualize patterns in the given data.

From the correlation matrix we can say:

1. High positive correlation between energy and loudness (+0.78) [which determines that with greater loudness in music, the energy increases]
2. High negative correlation between  energy and acousticness (-0.74) [which determines that with greater acoustiness, energy is decreased]
3. Fairly high positive correlation between danceability and valance (+0.56) [Valence is sound positiveness, the more valence, the danceability of the music increases]
4. Failry high negative correlation between loudness and acousticness (-0.56) [which determines that with greater acoustiness, loudness is decreased]

Note:
a. Correlation coefficient along the diagonal of the table are equal to 1 because each are perfectly correlated with itself.

```python
fig,ax = plt.subplots(4,3,figsize=(15,20))
sns.set(font_scale = 1)
sns.distplot(df['valence'],ax=ax[0,0])
sns.distplot(df['tempo'],ax=ax[0,1])
sns.distplot(df['acousticness'],ax=ax[0,2])
sns.distplot(df['danceability'],ax=ax[1,0])
sns.distplot(df['speechiness'],ax=ax[1,1])
sns.distplot(df['mode'],ax=ax[1,2])
sns.distplot(df['liveness'],ax=ax[2,0])
sns.distplot(df['loudness'],ax=ax[2,1])
sns.distplot(df['popularity'],ax=ax[2,2])
sns.distplot(df['energy'],ax=ax[3,0])
sns.distplot(df['year'],ax=ax[3,1])
sns.distplot(df['key'],ax=ax[3,2])
```

The above plot shows in visualization the distribution of each attributes of music

```python
number_of_releases = pd.DataFrame(df['year'].value_counts())
number_of_releases.rename({'year': 'releases'}, axis=1, inplace=True)
number_of_releases = number_of_releases.sort_index()
sns.set(font_scale=2) 
ax=number_of_releases.plot(kind='line',figsize=(15,8), linewidth=2)
plt.title("Number of songs released per year on spotify",y=1.05,fontsize=20)
plt.xlabel('Year')
plt.ylabel('Count of Songs')
ax.axes.get_xaxis().set_visible(True)
```

The above graph displays the number of songs released from the year 1930 to 2020 on Spotify

```python
top_ten_artists = df.groupby("artists")['popularity'].mean().sort_values(ascending=False).head(10)
top_ten_artists.head(10)

top_ten_artists.plot (kind = 'bar', figsize = (10,5))
```

The above graph displays the top 10 artists based on the popularity in each year.

```python
top_ten_tracks = df.groupby("name")['popularity'].mean().sort_values(ascending=False).head(10)
top_ten_tracks.head(10)

top_ten_tracks.plot(kind = 'bar', figsize = (15,8))
```

The above graph displays the top songs based on popularity over the years

```python

plt.figure(figsize=(15,5))
g=sns.barplot(x='Song Decade', y= 'duration_min', data=df, ci=True, palette='viridis')
g.bar_label(g.containers[0], fmt='%.1f', fontsize=14)
plt.ylabel('Length (minutes)', fontsize=14)
plt.xticks(fontsize=18)
plt.xlabel('Release Year', fontsize=18)
plt.title('Length vs Release Year', fontsize=20)
plt.show()
```

The above graph displays the length or duration of songs over the years

```python
plt.figure(figsize=(15,5))
g=sns.barplot(x='Song Decade', y= 'loudness', data=df, ci=False, palette='Reds_r')
g.bar_label(g.containers[0], fmt='%.1f', fontsize=14)
plt.ylabel('Loudness', fontsize=14)
plt.xticks(fontsize=18)
plt.xlabel('Release Year', fontsize=18)
plt.title('Loudness vs Release Year', fontsize=20)
plt.show()
```

The above graph shows the loudness variability in each year

```python
plt.figure(figsize=(20,10))
sns.regplot(data = df,x='energy',y='loudness',color='c').set(title='loudness vs energy correlation')
```

The above graph shows correlation between loudness and energy
